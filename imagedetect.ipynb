```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(imgnis2,cv2.COLOR_BGR2GRAY)
```

```python
faces = faceCascade.detectMultiScale(gray,1.1,4)
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(imgnis2,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
cv2.putText(imgnis2,
           predictionsnis2['dominant_emotion'],
            (0,50),
            font,1,
            (0,0,255),
            2,
            cv2.LINE_4
           )
```
